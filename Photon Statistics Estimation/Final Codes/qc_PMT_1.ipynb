```python
# New Section
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


```

```python
df = pd.read_csv('abc 100 1 ms 1 Lakhs pmt on.csv',  names=['y','x'])
df.head()
sample=df
np_X=np.array(sample['x'])
plt.hist(np_X,bins=[0,10,20,30,40,50,60,70])
```

```python
w=10.0
while(w>=1):
  plt.hist(np_X, edgecolor='black', bins=np.arange(min(np_X), max(np_X) + w, w))
  plt.title("Histogram with bin-size = %i " %w)
  plt.show()
  w=w/2
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values
x = np.linspace(0, 5, 100)  # Creating 100 points from 0 to 5

# Calculate y values
y = np.exp(-x)

# Create the plot
plt.plot(x, y)

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of y = e^(-x)')

# Show the plot
plt.grid(True)
plt.show()
```
